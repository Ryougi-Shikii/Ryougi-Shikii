```python

import tensorflow as tf
(train, train_label), (test, test_label) = tf.keras.datasets.fashion_mnist.load_data()

train = train.reshape(60000, 28 ,28 ,1)
test = test.reshape(10000, 28, 28, 1)

train = train / 255.0
test = test / 255.0
```

```python
print(train.shape)
print(test.shape)
```

```python

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3)),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation=tf.nn.relu),
    tf.keras.layers.Dense(units=64, activation=tf.nn.relu),
    tf.keras.layers.Dense(units=10, activation=tf.nn.sigmoid)
])
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
model.fit(train, train_label, epochs = 10)

test_loss, test_accuracy = model.evaluate(test, test_label)
print ('Test loss: {}, Test accuracy: {}'.format(test_loss, test_accuracy*100))
```
